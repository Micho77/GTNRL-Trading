**NOTE** Here the portfolio restriction is l1 norm = 1, not sum(w)=1

```python
from sklearn.metrics import pairwise_distances
import pandas as pd
import numpy as np
```

```python
def portfolio_performance(my_rs):
    
    daily_rs = round(my_rs.mean(), 5)
    daily_vol = round(my_rs.std(), 5)
    daily_sharpe = round(daily_rs/daily_vol, 5)
    
    current_dd = 0
    max_dd = 0
    for n in -my_rs:
        current_dd = max(current_dd+n, 0)
        max_dd = max(max_dd, current_dd)
    max_dd = round(max_dd, 5)
    
    return daily_rs, daily_vol, daily_sharpe, max_dd
```

```python
data_ohlc = pd.read_excel('FX Risk Premia Data.xlsx', 'OHLC', index_col=0, header=[3,5]).unstack()
prices = data_ohlc.loc[:, 'PX_LAST'].unstack().T
prices.sort_index(inplace=True)
prices = prices.astype(float)
prices.columns = ['AUDUSD', 'USDCAD', 'USDCHF', 'EURUSD', 'GBPUSD', 'USDJPY', 'USDNOK', 'NZDUSD', 'USDSEK']

prices.head(2)
```

```python
crs = pd.read_excel('FX Risk Premia Data.xlsx', 'CR', index_col=0, header=3).iloc[3:]
crs.sort_index(inplace=True)
crs = crs.astype(float)
crs.columns = ['EURUSD', 'GBPUSD', 'USDJPY', 'AUDUSD', 'NZDUSD', 'USDCAD', 'USDCHF', 'USDNOK', 'USDSEK']

crs.head(2)
```

```python
forwards = pd.read_excel('FX Risk Premia Data.xlsx', '1MF', index_col=0, header=3).iloc[3:]
forwards.sort_index(inplace=True)
forwards = forwards.astype(float)
forwards.columns = ['EURUSD', 'GBPUSD', 'USDJPY', 'AUDUSD', 'NZDUSD', 'USDCAD', 'USDCHF', 'USDNOK', 'USDSEK']

forwards.head(2)
```

# Classic Carry Strategy

```python
# Adjust implied forward points
forwards['USDJPY']*=100
```

```python
carry_score = forwards/prices
carry_score.plot(figsize=(15,3), legend=0, linewidth=2, grid=True)
```

```python
rs = crs.apply(np.log).diff(1).fillna(0)
```

```python
pos = carry_score.ffill().rank(1)
pos -= pos.median(1).values.reshape(-1,1)
pos[abs(pos)<=1] = 0
pos = pos.apply(np.sign)*-1
pos /= pos.abs().sum(1).values.reshape(-1,1)
pos = pos.shift(1)
```

```python
myrs = (rs*pos).sum(1)
myrs.apply(np.exp).cumprod().plot(figsize=(15,3), linewidth=2, grid=True)
```

```python
r, v, s, d = portfolio_performance(myrs)
print('rs: {}, vol: {}, sharpe: {}, max dd: {}'.format(r,v,s,d))
```

# Classic Momentum Strategy

**Using Carry Returns X-Currency Momentum**

```python
w = 252

rs = crs.apply(np.log).diff(1).fillna(0)
pos = rs.rolling(w).sum().rank(1).fillna(0)

pos -= pos.median(1).values.reshape(-1,1)
pos[abs(pos)<=1] = 0
pos = pos.apply(np.sign)
pos /= pos.abs().sum(1).values.reshape(-1,1)
pos = pos.shift(1)
```

```python
rs = crs.apply(np.log).diff(1).fillna(0)
myrs = (rs*pos).sum(1)
myrs.apply(np.exp).cumprod().plot(figsize=(15,3), linewidth=2, grid=True)
```

```python
r, v, s, d = portfolio_performance(myrs)
print('rs: {}, vol: {}, sharpe: {}, max dd: {}'.format(r,v,s,d))
```

# Laplacian Eigen-Vector Momentum Strategy

**<font color=red> (1) Are eigen-vectors defined for asymmetric matrices? (2) Largest or smallest eigen-vector?**


**Adaptive method**

```python
def pairwise_outperformance (rs1, rs2):
    return sum(list(rs1))-sum(list(rs2))
```

```python
def comput_w_opt (past_rs):
    
    # Pairwise out-performance
    A = pd.DataFrame(data=pairwise_distances(past_rs.T, metric=pairwise_outperformance),
                     index=rs.columns,
                     columns=rs.columns)
    
    # Clip negative values 
    A = A.clip(0, np.inf)
    
    # Eigen-decomposition
    e_vals, e_vecs = np.linalg.eigh(A)
    
    # Portfolio setup
    w_opt = pd.Series(e_vecs[:, -1], index=past_rs.columns)
    w_opt = np.sign(w_opt)
    w_opt[abs(w_opt)!=1] = 0
    w_opt /= sum(abs(w_opt))
    
    return w_opt
```

```python
w = 252 # look back window
skip = 22 # rebalancing days

pos = pd.DataFrame(index=rs.index, columns=rs.columns, dtype=float)
for t in rs.index[w:][[i for i in range(0, rs.shape[0]-w, skip)]]:
    # print(t)
    past_rs = rs.loc[:t].iloc[-w:]
    pos.loc[t] = comput_w_opt(past_rs)
```

```python
myrs = (pos.ffill().shift(1)*rs).sum(1)
myrs.apply(np.exp).cumprod().plot(figsize=(15,3), linewidth=2, grid=True)
```

```python
r, v, s, d = portfolio_performance(myrs)
print('rs: {}, vol: {}, sharpe: {}, max dd: {}'.format(r,v,s,d))
```

# Carry Graph Smoothed Momentum

**<font color=red> How should graph adjacency matrix be constructed (i.e. which direction)?**
    
How to determine the coefficient?

```python
def pairwise_carry_score (cs1, cs2):
    return sum(cs1-cs2)
```

```python
def filter_returns (past_rs, past_forwards):

    coef = 0.05

    A = pd.DataFrame(data=pairwise_distances(past_forwards.ffill().bfill().T, metric=pairwise_carry_score),
                     index=rs.columns,
                     columns=rs.columns)

    A = A.clip(0, np.inf)
    D_vals = A.sum(1)
    A = pd.DataFrame(data=np.diag(1/D_vals**0.5).dot(A).dot(np.diag(1/D_vals**0.5)),
                     index=A.index,
                     columns=A.columns).replace([np.nan, np.inf], 0)

    F = ((1-coef)*np.eye(A.shape[0]) + coef*A) 
    
    return past_rs.dot(F.T)
```

```python
w = 252 # look back window
skip = 22 # rebalancing days

rs = crs.apply(np.log).diff(1).fillna(0)
pos = pd.DataFrame(index=rs.index, columns=rs.columns, dtype=float)
for t in rs.index[w:][[i for i in range(0, rs.shape[0]-w, skip)]]:
    # print(t)
    past_rs = rs.loc[:t].iloc[-w:]
    past_forwards = forwards.loc[:t].iloc[-w:]
    
    filtered_rs = filter_returns (past_rs, past_forwards)
    pos.loc[t] = comput_w_opt(filtered_rs)
```

```python
rs = crs.apply(np.log).diff(1).fillna(0)
myrs = (pos.ffill().shift(1)*rs).sum(1)
myrs.apply(np.exp).cumprod().plot(figsize=(15,3), linewidth=2, grid=True)
```

```python
r, v, s, d = portfolio_performance(myrs)
print('rs: {}, vol: {}, sharpe: {}, max dd: {}'.format(r,v,s,d))
```
