# 3. Deep Learning Setup

```python
import tensorflow as tf
from tensorflow.keras.regularizers import l2
```

## 3.1. Model Definition

```python
def get_strategy_model (input_f_shape=(20, 42, ), current_w_shape=(6, )):

    ''' Policy gradient reinforcement learning strategy model.
    '''

    # Define two sets of inputs =========================================================================
    input_f = tf.keras.layers.Input(input_f_shape) # input features 
    current_w = tf.keras.layers.Input(current_w_shape) # current portfolio weights

    # features generation model =========================================================================
    fg1 = tf.keras.layers.GRU(32, activation='tanh', return_sequences=True)(input_f) # features generation 1
    fg2 = tf.keras.layers.GRU(32, activation='tanh', return_sequences=False)(fg1) # features generation 2
    fg2 = tf.keras.layers.Dense(32, activation='sigmoid')(fg2) # features generation 2

    model_fg = tf.keras.models.Model(inputs=input_f, outputs=fg2) # set up features generations model

    # strategy model ====================================================================================
    strategy_input = tf.keras.layers.concatenate([fg2, current_w]) # features and current weights
    sfg1 = tf.keras.layers.Dense(32, activation='sigmoid')(strategy_input) # strategy features generation 1
    w = tf.keras.layers.Dense(6, activation='tanh')(sfg1) # output weights

    return tf.keras.models.Model(inputs=[input_f, current_w], outputs=w)
```

## 3.2. Loss Definition

```python
def strategy_loss (w_current, tc=0):
    
    def penalized_loss(y_true, y_pred):
    
        ''' Computes the negative Sharpe ratio generated during the period covered by the batch.

        Future work:
            - Include non zero transaction costs.
            - Set positions to l1norm = 1.
            - Include a boolean mask to set not high enough predictions to 0 in position.
            - Add similarity penalty to the weights of VAR strategy

        '''
        
        my_tc = tc * tf.math.reduce_sum(abs(y_pred - w_current), 1) # penalty on portfolio change
        my_rs = tf.reduce_sum(y_true*y_pred, axis=1) - my_tc # portfolio sum of element wise returns - costs
        sharpe = tf.math.reduce_mean(my_rs)/tf.math.reduce_std(my_rs) # compute sharpe ratio

        return -1*sharpe
    
    return penalized_loss
```

# 4. Pseudo-RL Training

Note: 
- When over-fit on transaction costs of 0.0002, the model is able to generate a profitable strategy that doesn't generalize
- When fit on transaction costs of 0.0002 for multiple periods, the model can't generate a profitable strategy

Idea:
- Fit on transaction costs of 0 for multiple periods to generate a profitable strategy that generalize well
- Fit a second time by introducing transaction costs using proper RL

```python
def sample_trajectory_period (X_tr, y_tr, duration):
    
    ''' Sample a random period in time. 
    '''
    
    total_steps = X_tr.shape[0]
    rnd_idx = np.random.randint(low=duration, high=total_steps)
    
    x_traj = X_tr[rnd_idx-duration:rnd_idx]
    y_traj = y_tr[rnd_idx-duration:rnd_idx]
    # y_traj = y_traj - y_traj.mean(0).reshape(1,-1) # subtract trend to avoid fixed weights strategy
    
    return x_traj, y_traj
```

```python
def print_performance (w_traj, y_traj, tc=0.0002):
    
    clear_output()
    my_ws = pd.DataFrame(w_traj)
    my_rs = (my_ws*y_traj).sum(1)
    my_cs = my_ws.diff(1).abs().sum(1) * tc
    my_rswcs = my_rs - my_cs

    fig, ax = plt.subplots(1,2, figsize=(15,2))
    my_rs.cumsum().plot(ax=ax[0], grid=True, linewidth=2, color='blue', title='no transaction costs')
    my_rswcs.cumsum().plot(ax=ax[1], grid=True, linewidth=2, color='blue', title='with transaction costs')
    plt.show()
    
    plt.figure(figsize=(15,2))
    plt.plot(w_traj)
    plt.show()
    
    pass
```

## 4.2. Sparse Sampling Training over Random Trajectories

**NO TRANSACTION COSTS TRAINING**: Fit directly on trading signals.

```python
def train_on_traj (mdl, opt, x_traj, y_traj, epochs1=30, epochs2=100):

    # Train over multiple instances of random initialization (prevent fitting on the given random weights)
    for e in range(epochs1):

        # generate random portfolio weights for the trajectory
        w_tr = np.random.random(size=(x_traj.shape[0], 6))-0.5 
        w_tr /= abs(w_tr).sum(1).reshape(-1,1)

        # Batch train strategy (need high epochs here, otherwise it takes forever)
        mdl.compile(loss=strategy_loss(w_tr, tc=0), optimizer=opt)
        mdl.fit([x_traj, w_tr], y_traj, epochs=epochs2, batch_size=x_traj.shape[0], verbose=0, shuffle=0)

        # print(e, mdl.history.history['loss'][-1])
        
    return mdl, mdl.history.history['loss'][-1]
```

```python
# Deep learning training setup
mdl = get_strategy_model()
opt = tf.keras.optimizers.Adam(lr=1e-3)

# Training variables
tr_scores = []
epochs = 1500
```

```python
# Train over multiple number of trajectories
# mdl.load_weights("model.h5")
for e in range(epochs):    

    # Randomly sample a period of N bars (note: massively unstable for high batchsize/duration)
    x_traj, y_traj = sample_trajectory_period (X_tr, y_tr, duration=1000)  
    
    # Fit on trajectory
    mdl, tr_score = train_on_traj (mdl, opt, x_traj, y_traj, epochs1=1, epochs2=1)
    tr_scores.append(tr_score)
    print(e, tr_score)

    # save weights
    mdl.save_weights("model.h5")
```

```python
pd.Series(tr_scores).plot(figsize=(15,3), logy=False)
pd.Series(tr_scores).rolling(20).mean().plot(figsize=(15,3), logy=False, color='red', grid=True)
```

## 4.3. Backtest over Training Period

```python
def generate_trajectory_actions (mdl, x_traj, y_traj):
    
    ''' Generate the strategy actions according to the model for the given period (trajectory).
    '''

    # initial portfolio weights
    w_t = np.zeros((1, y_traj.shape[1]))
    w_t = w_t.reshape(1, -1)

    # generate trajectory
    w_traj = []
    for i in range(x_traj.shape[0]):
        x_t = x_traj[i].reshape(1, 20, -1)
        w_t = mdl.predict([x_t, w_t])      
        w_traj.append(w_t.reshape(-1))
        
    return np.array(w_traj)
```

```python
# Randomly sample a period of N bars 
x_traj, y_traj = sample_trajectory_period (X_tr, y_tr, duration=10000)  

plt.figure(figsize=(15,2))
plt.plot(y_traj.cumsum(0))
plt.show()
```

```python
w_traj = generate_trajectory_actions (mdl, x_traj, y_traj)
print_performance (w_traj, y_traj, tc=0.0002)
```
